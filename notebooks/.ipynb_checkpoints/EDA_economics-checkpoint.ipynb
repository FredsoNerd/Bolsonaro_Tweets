```python
# Useful parameters for the code
initialDate = '2008-01-01'
indices = list(df.columns)[1:]
nRows = int(len(indices)/2)

# Y Label, Graph title
dicTitles = {'ipca':['Monthly Percentage Value','IPCA'],
 'igpm':['Monthly Percentage Value','IGP-M'],
 'inpc':['Monthly Percentage Value','INPC'],
 'selic_meta':['Yearly Percentage Value','SELIC Meta'],
 'international_reserve':['Total daily value (millions of USD)','International Reserve'],
 'pnad':['Percentage Value','PNAD'],
 'cdi':['Daily Percentage Value','CDI'],
 'gdp':['Monthly Values (millions of R$)','GDP'],
 'dollar':['Brazilian Reais','Dollar Exchange'],
 'employment':['Index Value','Formal Employment'],
 'gov_debt':['Liquid Debt of Public Sector (%GDP)','Governamental Debt'],
 'consumer_confidence':['Index Value','Consumer Confidence']
}

# Plot
fig, ax = plt.subplots(nRows, 2, figsize=(14,6*nRows))

for i in range(0,len(indices)):
    # Getting the info
    allValues = np.array(df[indices[i]])
    ind = ~np.isnan(allValues) & (np.array(df['data'])>=initialDate)
    seriesValues = allValues[ind]
    dateValues = pd.to_datetime(df['data'][ind])
    
    # Plotting and formatting
    ax[int(i/2)][i%2].plot(dateValues, seriesValues, lw=0.8)
    
    # Set axes limits
    ax[int(i/2)][i%2].set_xlim(dateValues.min(), dateValues.max())

    # Make labels
    ax[int(i/2)][i%2].set_xlabel('Year', fontsize=14)
    if 'Percentage Value' in dicTitles[indices[i]][0]:
        ax[int(i/2)][i%2].set_ylabel(dicTitles[indices[i]][0], fontsize=14)
        ax[int(i/2)][i%2].yaxis.set_major_formatter(matplotlib.ticker.PercentFormatter(100))
    else:
        ax[int(i/2)][i%2].set_ylabel(dicTitles[indices[i]][0], fontsize=14)
        
    ax[int(i/2)][i%2].set_title(dicTitles[indices[i]][1], fontsize=18)
    
    # Grid
    ax[int(i/2)][i%2].grid(True, lw=1.5, ls='--', alpha=0.75)
    
# Overall title
plt.suptitle("Variation of diverse economic indexes",x=0.5, y=0.9,verticalalignment='top', fontsize = 20)
plt.show()
```
